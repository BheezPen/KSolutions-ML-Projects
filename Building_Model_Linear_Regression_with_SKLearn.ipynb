```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.metrics import r2_score
```

```python
raw_data = pd.read_csv('olympics.csv')

raw_data
```

```python
#Instantating the model
my_model = LinearRegression()

#The my_model should then follow the syntax

#model.fit('independent variable', 'dependent variable')
#where
#x == independent variable
#y == dependent variable
```

```python
# TO extract the x and y
x = raw_data[['athletes','prev_medals']].copy()

x 
```

```python
#Now that i have the x copied, next is that i pass the categorical value into the x
# Remember, when i was writing the algorithm manually, i added unit coloumn (i.e 1) all through
# i.e i added 1 all through as another colummn

x['intercepts'] = 1

x
```

```python
# After adding that, let me re arrange such that the intercept comes first
x  = x[['intercepts','athletes','prev_medals']]
x #SO now this is the x value, (i.e independent variable) we will be passing into the model
```

```python
y = raw_data[['medals']].copy()
y #ANd now this y is the dependent variable to be passed into the model
```

```python
# NOw, lets pass into the following the syntex below
# model.fit('independent variable', 'dependent variable')
#Since, my instantiation (i.e object) is my_model, then

my_model.fit(x,y)
```

```python
# Now lets calculate the coefficient value.
# Remember the coefficient value (i.e B) as used when calculating manually
coefficient_value = my_model.coef_

coefficient_value
```

```python
# After getting the value for coefficinet value (B)
#Next is to calculate for yhat, i.e estimated Y, aka Prediction

prediction = my_model.predict(x)
# Recall, prediction == yHat from my manual math here
prediction
```

```python
#Now next is to calculate the rsquare
# Becuase we are not caltulating manually, no need to solve for ssr and sst
# We only import the rsquare to help check the accuracy metrics
# we then say
# from sklearn.metrics import r2_score which has been done int he first cell

r2_score(y,prediction)
```
